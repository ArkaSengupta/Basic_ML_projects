```python
plt.subplots(figsize = (8,8))
sns.heatmap(df.corr(),annot=True,fmt="f").set_title("Corelation of attributes among df")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import train_test_split

X = df.iloc[:,0:10].values
y = df.iloc[:,10].values

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)

model = LinearRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
# print('mean_squared_error : ', round(mean_squared_error(y_test, predictions), 2))
print('mean_absolute_error : ', round(mean_absolute_error(y_test, predictions), 2))
print('mean_absolute_error_percent : ', round(mean_absolute_error(y_test, predictions) / np.mean(y_test), 2), '%' )
print('Accuracy : ', round(100 - mean_absolute_error(y_test, predictions) / np.mean(y_test), 2), '%' )
```
